```python
# basic functions and manipulations

import cv2 as cv
import numpy as np


img = cv.imread('resources\img.png')

# resize
resized_img = cv.resize(img,(450,250))

# gray
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# blurred img
b_img = cv.GaussianBlur(img, (13,13), 0)

# b_W
thresh , binary = cv.threshold(gray, 117,215, cv.THRESH_BINARY)

# edge detection 
edge_img = cv.Canny(img, 48,48)


# thickness of lines
mat_kernel = np.ones((3,3), np.int8)
dilated_img = cv.dilate(edge_img, (7,7), iterations=1)
dilated_img = cv.dilate(edge_img, (mat_kernel), iterations=1)


# make thinner outline
ero_img = cv.erode(dilated_img, mat_kernel, iterations=1)

# crop - using numpy
print(img.shape)
cropped_img = img[0:200, 0:300]











cv.imshow('original', img)
cv.imshow('r', resized_img)
cv.imshow('gray', gray)
cv.imshow('b', b_img)
cv.imshow('e',edge_img)
cv.imshow('d', dilated_img)
cv.imshow('e', ero_img)
cv.imshow('b_w', binary)
cv.imshow('o', img)
cv.imshow('c', cropped_img)

cv.waitKey(0)
 

cv.destroyAllWindows()
```

```python

# changing perspective of image


import cv2 as cv
import numpy as np

img = cv.imread('resources/img.png')
print(img.shape)
# defining points
point1 = np.float32([[233,196],[82,471],[522,169],[715,482]])

width , height = 800,900

point2 = np.float32([[0,0],[800,0],[0,height],[width,height]])

matrix = cv.getPerspectiveTransform(point1, point2)
out_img = cv.warpPerspective(img,matrix,(width,height))

cv.imshow('t', out_img)
cv.waitKey(0)
cv.destroyAllWindows()
 
```

```python

# finding coordinates and color codes of imgs

import cv2 as cv
import numpy as np

# defing func

def find_coord(event , x, y, flags, params):
    if event == cv.EVENT_LBUTTONDOWN:
    # on left mouse click
        print(x , '', y)
    # how to define or print the same image or window
        font = cv.FONT_HERSHEY_PLAIN
        cv.putText(img, str(x)+','+str(y),(x,y),font,1, (255,0,0), thickness=2)
    # show the text on image and img
        cv.imshow('img',img)

    # for color finding
    if event == cv.EVENT_RBUTTONDOWN:
        print(x , ',', y)

        font = cv.FONT_HERSHEY_SIMPLEX

        b = img[y,x,0]
        g = img[y,x,1]
        r = img[y,x,2]

        cv.putText(img, str(b)+','+str(g)+','+str(r),(x,y),font, 1, (255,0,0), 2)
        cv.imshow('img',img)



# final function to read and display

if __name__=='__main__':
    #reading image
    img = cv.imread('img.png')
    #display img
    cv.imshow("img", img)
    # setting call back function
    cv.setMouseCallback('img', find_coord)

    cv.waitKey(0)
    cv.destroyAllWindows()


```

```python
# how to detect specific colors inside python

import cv2 as cv
import numpy as np


#img = cv.imread('img.png')

# convert in hsv (Hue, Saturation, value)

#hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)



#sliders
def slider():
    pass

path = "img.png"

cv.namedWindow('bars')    # creating new window
cv.resizeWindow("bars",700,300)

cv.createTrackbar("hue min",'bars', 0,179, slider)   # adding sliders
cv.createTrackbar('hue max','bars',179,179,slider)

cv.createTrackbar("sat min",'bars', 0,255, slider)
cv.createTrackbar('sat max','bars',255,255,slider)

cv.createTrackbar("val min",'bars', 0,255, slider)
cv.createTrackbar('val max','bars',255,255,slider)




while True:
    img = cv.imread(path)
    hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)     # hsv conversion
    hue_min = cv.getTrackbarPos('hue min', 'bars')    
    hue_max = cv.getTrackbarPos('hue max', 'bars')
    sat_min = cv.getTrackbarPos('sat min', 'bars')
    sat_max = cv.getTrackbarPos('sat max', 'bars')
    val_min = cv.getTrackbarPos('val min', 'bars')
    val_max = cv.getTrackbarPos('val max', 'bars')
    print(hue_min,hue_max,sat_min,sat_max,val_min,val_max)
    
    # to see these inside imgs
    lower = np.array([hue_min,sat_min,val_min])
    upper = np.array([hue_max,sat_max,val_max])
    mask_img = cv.inRange(hsv_img,lower,upper)
    out_img = cv.bitwise_and(img,img, mask=mask_img)

    cv.imshow('o', img)
    cv.imshow('h', hsv_img)
    cv.imshow('m',mask_img)
    cv.imshow('o',out_img)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break 
cv.destroyAllWindows()

```

```python
# how to detect face in imgs

import cv2 as cv
import numpy as np


face_cascade = cv.CascadeClassifier('C:/Users/Gulfam/Desktop/Data Sci/opencv/haarcascade_frontalface_default.xml')

img = cv.imread('860_main_beauty.png')

#img = cv.resize(img,(683,384))

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray_img, 1.1,4)

# draw a rectangle
for (x,y,w,h) in faces:
    cv.rectangle(img, (x,y), (x+w, y+h),(255,0,0),2)


cv.imshow('img',img)
cv.waitKey(0)
cv.imwrite('facedetection.png',img)
cv.destroyAllWindows()
     
```
